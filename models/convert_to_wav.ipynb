# Defining a Function for MP3 to WAV Conversion

```python
import librosa
import os
import soundfile as sf
```

```python
def convert_to_wav(audio_file):
    audio,sr=librosa.load(audio_file)
    wav_file=os.path.splitext(audio_file)[0]+".wav"
    sf.write(wav_file,audio,sr)
    print(f"Converted {audio_file} to {wav_file}")
```

```python
convert_to_wav("afrikaans5.mp3")
```

# Converting All MP3 Files in a Folder

```python
def convert_folder_to_wav(input_folder,output_folder):
    os.makedirs(output_folder,exist_ok=True)
    for filename in os.listdir(input_folder):
        if filename.lower().endswith(".mp3"):
            audio_file=os.path.join(input_folder, filename)
            base_name=os.path.splitext(filename)[0]
            wav_file=os.path.join(output_folder,base_name+".wav")
            audio,sr=librosa.load(audio_file)
            sf.write(wav_file,audio,sr)
    print("All MP3 files in the folder have been converted to WAV.")
```

```python
convert_folder_to_wav("recordings","Output_Folder")
```
