```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
```

```python
audio,sr=librosa.load("Demo1.wav")
```

```python
Audio(data=audio,rate=sr)
```

```python
plt.figure(figsize=(12,4))
librosa.display.waveshow(audio,sr=sr)
plt.title('Audio Waveform')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
# Remove silence
audio_trimmed,_=librosa.effects.trim(audio,top_db=20) # removes the silence at the start and end of the audio
Audio(data=audio_trimmed,rate=sr)
```

```python
plt.figure(figsize=(12,4))
librosa.display.waveshow(audio_trimmed,sr=sr)
plt.title('Trimmed Audio Waveform')
plt.xlabel('Time')
plt.ylabel('Amplitude
```

```python
# Apply pre-emphasis filter (enhances high frequencies)
pre_emphasis = 0.97
audio_emphasized=np.append(audio_trimmed[0],audio_trimmed[1:]-pre_emphasis*audio_trimmed[:-1])
```

```python
Audio(data=audio_emphasized,rate=sr)
```

```python
# Normalize amplitude
audio_normalized=librosa.util.normalize(audio_emphasized)
```

```python
Audio(data=audio_normalized,rate=sr)
```
